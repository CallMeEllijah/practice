### Get all the sources that newsAPI has

### Get news


```python
all_articles = newsapi.get_everything(sources='google-news',
                                      from_param='2021-03-11',
                                      to='2021-03-12')
```

### Save JSON results

```python
with open('news.json', 'w') as outfile:
  json.dump(all_articles, outfile)
files.download('news.json')
```
